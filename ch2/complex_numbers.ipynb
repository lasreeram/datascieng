## Real-Root Polynomial Examples

We start with two examples that have real roots:

1. A degree-2 polynomial (quadratic)
2. A degree-3 polynomial (cubic)

These examples motivate how roots appear as x-axis intersections in real-valued plots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Degree-2 example: p(x) = x^2 - 1 (real roots at x = -1, 1)
x = np.linspace(-3, 3, 600)
p2 = x**2 - 1

plt.figure(figsize=(7, 4))
plt.plot(x, p2, linewidth=2, label=r"$p(x)=x^2-1$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.scatter([-1, 1], [0, 0], color="red", zorder=3, label="real roots")
plt.title("Degree 2 Polynomial with Real Roots")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

For the cubic example, choose
$$
q(x) = (x+2)(x-1)(x-3).
$$
This is degree 3 and has three real roots: $x=-2,1,3$.

```python
# Degree-3 example: q(x) = (x+2)(x-1)(x-3)
x = np.linspace(-4, 4, 800)
q = (x + 2) * (x - 1) * (x - 3)

plt.figure(figsize=(7, 4))
plt.plot(x, q, linewidth=2, color="purple", label=r"$q(x)=(x+2)(x-1)(x-3)$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.scatter([-2, 1, 3], [0, 0, 0], color="red", zorder=3, label="real roots")
plt.title("Degree 3 Polynomial with Real Roots")
plt.xlabel("x")
plt.ylabel("q(x)")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

Now add higher-degree examples with real roots:

- Degree 4: $$r(x)=(x+3)(x+1)(x-1)(x-2)$$
- Degree 5: $$s(x)=(x+2)(x+1)x(x-1)(x-3)$$

```python
x = np.linspace(-4, 4, 1200)

# Degree-4 polynomial
r = (x + 3) * (x + 1) * (x - 1) * (x - 2)
r_roots = [-3, -1, 1, 2]

plt.figure(figsize=(7, 4))
plt.plot(x, r, linewidth=2, color="teal", label=r"$r(x)=(x+3)(x+1)(x-1)(x-2)$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.scatter(r_roots, [0] * len(r_roots), color="red", zorder=3, label="real roots")
plt.title("Degree 4 Polynomial with Real Roots")
plt.xlabel("x")
plt.ylabel("r(x)")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

Now plot the degree-5 polynomial on a separate graph with more widely spaced roots:

$$
s(x)=0.02(x+5)(x+2)(x-1)(x-4)(x-9)
$$

The real roots are at $x=-5,-2,1,4,9$.

```python
x = np.linspace(-6, 10, 1400)

# Degree-5 polynomial with more widely spaced roots
s = 0.02 * (x + 5) * (x + 2) * (x - 1) * (x - 4) * (x - 9)
s_roots = [-5, -2, 1, 4, 9]

plt.figure(figsize=(8, 4.5))
plt.plot(x, s, linewidth=2, color="darkorange", label=r"$s(x)=0.02(x+5)(x+2)(x-1)(x-4)(x-9)$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.scatter(s_roots, [0] * len(s_roots), color="red", zorder=3, label="real roots")
plt.title("Degree 5 Polynomial with More Spaced Real Roots")
plt.xlabel("x")
plt.ylabel("s(x)")
plt.xlim(-6, 10)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

Example with **multiple (repeated) roots**, spaced farther apart:

$$
m(x)=0.001(x+5)^2(x-1)(x-8)^2
$$

- Repeated root at $x=-5$ (multiplicity 2)
- Simple root at $x=1$ (multiplicity 1)
- Repeated root at $x=8$ (multiplicity 2)

With wider spacing, the polynomial shape is easier to see between roots.

Example with an **odd multiplicity root**:

$$
o(x)=0.001(x+5)^2(x-1)^3(x-8)
$$

- Root at $x=1$ has multiplicity 3 (odd), so the graph crosses the x-axis with flattening.
- Root at $x=-5$ has multiplicity 2 (even), so it touches and turns.

Example with **no real roots** (real coefficients):

A real-coefficient cubic must have at least one real root, so we use degree 4:

$$
p(x)=x^4+1
$$

Since $x^4\ge 0$ for all real $x$, we have $x^4+1>0$ for all real $x$, so it never crosses the x-axis.

```python
x = np.linspace(-2.2, 2.2, 1000)

# Degree-4 polynomial with no real roots
p_no_real = x**4 + 1

plt.figure(figsize=(12, 6), dpi=150)
plt.plot(x, p_no_real, linewidth=3, color="darkgreen", label=r"$p(x)=x^4+1$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.title("Degree 4 Polynomial with No Real Roots", fontsize=20)
plt.xlabel("x", fontsize=14)
plt.ylabel("p(x)", fontsize=14)
plt.ylim(-0.5, 8)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

Another no-real-roots example:

$$
q(x)=x^4+4x^2+5

This is always positive because
$$
q(x)=(x^2+2)^2+1>0

for all real $x$.

```python
x = np.linspace(-3, 3, 1200)

# Another degree-4 polynomial with no real roots
q_no_real = x**4 + 4*x**2 + 5

plt.figure(figsize=(12, 6), dpi=150)
plt.plot(x, q_no_real, linewidth=3, color="darkslateblue", label=r"$q(x)=x^4+4x^2+5$")
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1, alpha=0.5)
plt.title("Another Degree 4 Polynomial with No Real Roots", fontsize=20)
plt.xlabel("x", fontsize=14)
plt.ylabel("q(x)", fontsize=14)
plt.ylim(-0.5, 22)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

Complex-plane view of
$$
q(z)=z^4+4z^2+5
$$

We plot $|q(z)|$ over $(\Re z, \Im z)$. Points where $|q(z)|=0$ are the complex roots.

```python
xr = np.linspace(-2.5, 2.5, 500)
yi = np.linspace(-2.5, 2.5, 500)
X, Y = np.meshgrid(xr, yi)

Z = X + 1j * Y
Qz = Z**4 + 4 * Z**2 + 5
abs_Qz = np.abs(Qz)

roots_q = np.roots([1, 0, 4, 0, 5])

plt.figure(figsize=(12, 7), dpi=150)
cont = plt.contourf(X, Y, abs_Qz, levels=80, cmap="viridis")
plt.colorbar(cont, label=r"$|q(z)|$")
plt.axhline(0, color="white", linewidth=1, alpha=0.8)
plt.axvline(0, color="white", linewidth=1, alpha=0.8)
plt.scatter(roots_q.real, roots_q.imag, color="red", s=70, label="complex roots")
plt.title(r"Complex Plane: $q(z)=z^4+4z^2+5$", fontsize=18)
plt.xlabel(r"Re$(z)$", fontsize=14)
plt.ylabel(r"Im$(z)$", fontsize=14)
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.grid(alpha=0.15)
plt.legend()
plt.tight_layout()
plt.show()
```

## Fundamental Theorem of Algebra

**Statement:** Every non-constant polynomial with complex coefficients has at least one complex root.

**Consequence:** A degree-$n$ polynomial has **exactly $n$ complex roots**, counted with multiplicity:

$$
p(z) = a_n z^n + a_{n-1}z^{n-1} + \cdots + a_0, \quad a_n \neq 0
$$

can always be fully factored as

$$
p(z) = a_n \prod_{k=1}^{n}(z - r_k)
$$

where each $r_k \in \mathbb{C}$ (roots may repeat).

**Key corollary for real polynomials:** If $p$ has real coefficients and $r = a+bi$ is a root, then its conjugate $\bar{r} = a-bi$ is also a root. So non-real roots come in conjugate pairs.

| Polynomial | Degree | Root locations |
|---|---|---|
| $(z-1)(z+2)(z-3)$ | 3 | All on real axis |
| $(z^2+1)(z^2+4)$ | 4 | All on imaginary axis |
| $z^4+1$ | 4 | Off both axes, conjugate pairs |

This is exactly what we see in the complex-plane magnitude plots above: the dark pits (zeros of $|p(z)|$) reveal all $n$ roots, wherever they lie.


## Roots of finite Taylor truncations of the exponential

For
$$
E_n(z)=\sum_{k=0}^{n}\frac{z^k}{k!},
$$
each finite $n$ gives a degree-$n$ polynomial, so it has exactly $n$ complex roots.

Below, we plot roots of $E_n(z)$ for several $n$ values to see how the pattern evolves as $n$ grows.


```python
import math

n_values = [5, 10, 15, 20, 30, 40]
colors = plt.cm.viridis(np.linspace(0.1, 0.95, len(n_values)))

plt.figure(figsize=(10, 8), dpi=150)

for n, color in zip(n_values, colors):
    coeffs = [1 / math.factorial(k) for k in range(n + 1)]
    roots_en = np.roots(coeffs[::-1])
    plt.scatter(roots_en.real, roots_en.imag, s=26, color=color, alpha=0.85, label=f"n={n}")

plt.axhline(0, color="black", linewidth=1, alpha=0.6)
plt.axvline(0, color="black", linewidth=1, alpha=0.6)
plt.title(r"Roots of $E_n(z)=\sum_{k=0}^{n} z^k/k!$", fontsize=16)
plt.xlabel(r"Re$(z)$", fontsize=13)
plt.ylabel(r"Im$(z)$", fontsize=13)
plt.grid(alpha=0.25)
plt.gca().set_aspect("equal", adjustable="box")
plt.legend(title="Truncation degree", fontsize=9)
plt.tight_layout()
plt.show()
```

## Why roots disappear in the limit

Let
$$
E_n(z)=\sum_{k=0}^{n}\frac{z^k}{k!}.
$$
Each $E_n$ is degree $n$, so it has exactly $n$ roots in $\mathbb{C}$ (counting multiplicity).

But
$$
E_n(z)\to e^z
$$
uniformly on every compact set, and
$$
e^z\neq 0\quad\text{for all }z\in\mathbb{C}.
$$

By Hurwitz's theorem: if $f_n\to f$ uniformly on compact sets and $f$ has no zeros in a domain, then zeros of $f_n$ cannot accumulate there.

So the zeros of $E_n$ cannot converge to any finite point; they must escape to infinity:
$$
|z_{n,j}|\to\infty\quad (n\to\infty).
$$

A simple analogy:
$$
p_n(z)=1+\frac{z}{n}\to 1,
$$
but each $p_n$ has a root at $z=-n$, which runs off to $-\infty$.

So there is no contradiction: finite truncations have roots, while the limit function $e^z$ has none because the roots drift outward as $n$ grows.


## Negative Numbers as a Phase Change — and Why Complex Numbers Arise Naturally

### Step 1: Multiplication as Rotation

On the number line, multiplying by $-1$ flips a number to the opposite side of zero:

$$1 \xrightarrow{\times(-1)} -1 \xrightarrow{\times(-1)} 1$$

Geometrically, this is a **rotation by $\pi$ radians** (180°) in the plane.

### Step 2: What is "half" of that rotation?

If $\times(-1)$ is a rotation by $\pi$, applying it **twice** returns to the start.  
What operator $j$ applied **twice** gives $-1$?

$$j \cdot j = -1 \implies j = \sqrt{-1} = i$$

This is a rotation by $\pi/2$ (90°). There is no real number satisfying this — so intermediate phases **force** a second dimension.

### Step 3: General Phase

Any rotation by angle $\theta$ can be written as multiplication by $e^{i\theta}$:

$$e^{i\theta} = \cos\theta + i\sin\theta$$

| $\theta$ | $e^{i\theta}$ | Geometric effect |
|----------|--------------|-----------------|
| $0$ | $1$ | Identity |
| $\pi/2$ | $i$ | Quarter turn |
| $\pi$ | $-1$ | Flip (negative numbers!) |
| $3\pi/2$ | $-i$ | Three-quarter turn |
| $2\pi$ | $1$ | Full rotation, back to start |

So **negative numbers are the special case $\theta=\pi$**. Complex numbers fill in all the intermediate rotations.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# ── Panel 1: Successive multiplication by -1 on the real line ──────────────
ax = axes[0]
for k, (p, c) in enumerate(zip([1, -1, 1, -1], ['tab:blue','tab:red','tab:blue','tab:red'])):
    ax.annotate('', xy=(p, 0), xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color=c, lw=2.2))
    ax.scatter(p, 0, color=c, s=80, zorder=5)
    ax.text(p, 0.1 * (-1)**k, f'step {k}: {p:+d}', ha='center', fontsize=9)
ax.axhline(0, color='gray', lw=0.8)
ax.axvline(0, color='gray', lw=0.8)
ax.set_xlim(-1.6, 1.6); ax.set_ylim(-0.5, 0.5)
ax.set_title('Multiplying by $-1$\n(rotation by $\\pi$)', fontsize=11)
ax.set_xlabel('Real axis')
ax.set_aspect('equal'); ax.grid(alpha=0.25)

# ── Panel 2: Successive multiplication by i ────────────────────────────────
ax = axes[1]
sequence = [1+0j, 1j, -1+0j, -1j, 1+0j]
labels   = ['$1$', '$i$', '$-1$', '$-i$', '$1$']
cols     = ['tab:blue','tab:green','tab:red','tab:purple','tab:blue']
unit_circle = np.exp(1j * np.linspace(0, 2*np.pi, 200))
ax.plot(unit_circle.real, unit_circle.imag, 'k--', lw=0.8, alpha=0.35)
for k in range(len(sequence)-1):
    z0, z1 = sequence[k], sequence[k+1]
    ax.annotate('', xy=(z1.real, z1.imag), xytext=(z0.real, z0.imag),
                arrowprops=dict(arrowstyle='->', color='gray', lw=1.5,
                                connectionstyle='arc3,rad=0.25'))
for z, lab, c in zip(sequence, labels, cols):
    ax.scatter(z.real, z.imag, color=c, s=90, zorder=5)
    ax.annotate(lab, (z.real, z.imag), textcoords='offset points',
                xytext=(8, 5), fontsize=10,
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))
ax.text(0, 0, r'$\times i = +90°$', ha='center', va='center', fontsize=8, color='gray')
ax.axhline(0, color='gray', lw=0.8); ax.axvline(0, color='gray', lw=0.8)
ax.set_xlim(-1.6, 1.6); ax.set_ylim(-1.6, 1.6)
ax.set_title('Multiplying by $i$\n(rotation by $\\pi/2$)', fontsize=11)
ax.set_xlabel('Real'); ax.set_ylabel('Imaginary')
ax.set_aspect('equal'); ax.grid(alpha=0.25)

# ── Panel 3: Continuous phase sweep e^{iθ} ────────────────────────────────
ax = axes[2]
theta = np.linspace(0, 2*np.pi, 500)
z = np.exp(1j * theta)
sc = ax.scatter(z.real, z.imag, c=theta, cmap='hsv', s=10)
plt.colorbar(sc, ax=ax, label='θ  (radians)')
specials = [(0, '$1$\n(θ=0)'), (np.pi/2, '$i$\n(θ=π/2)'),
            (np.pi, '$-1$\n(θ=π)'), (3*np.pi/2, '$-i$\n(θ=3π/2)')]
for ang, lab in specials:
    z0 = np.exp(1j * ang)
    ax.scatter(z0.real, z0.imag, color='black', s=70, zorder=5)
    ax.annotate(lab, (z0.real, z0.imag), textcoords='offset points',
                xytext=(10, 5), fontsize=9,
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))
ax.axhline(0, color='gray', lw=0.8); ax.axvline(0, color='gray', lw=0.8)
ax.set_xlim(-1.6, 1.6); ax.set_ylim(-1.6, 1.6)
ax.set_title('All phases $e^{i\\theta}$\nComplex numbers fill the circle', fontsize=11)
ax.set_xlabel('Real'); ax.set_ylabel('Imaginary')
ax.set_aspect('equal'); ax.grid(alpha=0.25)

plt.suptitle('Negative Numbers are Phase $\\pi$ — Complex Numbers are Intermediate Phases',
             fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```
