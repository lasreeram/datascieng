# Interactive Solver: System of 2 Linear Equations

Enter coefficients for
$$
a_1x + b_1y = c_1,
$$
$$
a_2x + b_2y = c_2.
$$
The graph updates automatically and shows the intersection point when it exists.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display


def solve_and_plot(a1, b1, c1, a2, b2, c2):
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot window and sample x-range
    x_min, x_max = -10, 10
    y_min, y_max = -10, 10
    x = np.linspace(x_min, x_max, 500)

    def plot_line(a, b, c, color, label):
        # Handle vertical line: ax + by = c with b = 0 => x = c / a
        if abs(b) < 1e-12:
            if abs(a) < 1e-12:
                return "invalid"
            x_const = c / a
            ax.axvline(x_const, color=color, linewidth=2, label=label + f"  (x={x_const:.3f})")
            return "vertical"
        y = (c - a * x) / b
        ax.plot(x, y, color=color, linewidth=2, label=label)
        return "normal"

    status1 = plot_line(a1, b1, c1, "tab:blue", "Equation 1")
    status2 = plot_line(a2, b2, c2, "tab:orange", "Equation 2")

    # Solve system using determinant
    det = a1 * b2 - a2 * b1

    if abs(det) > 1e-12:
        x_int = (c1 * b2 - c2 * b1) / det
        y_int = (a1 * c2 - a2 * c1) / det
        ax.scatter(x_int, y_int, color="red", s=90, zorder=5, label=f"Intersection ({x_int:.3f}, {y_int:.3f})")
        result_text = f"Unique intersection: x = {x_int:.6f}, y = {y_int:.6f}"
    else:
        # Parallel or coincident lines
        same_line = (
            abs(a1 * c2 - a2 * c1) < 1e-12
            and abs(b1 * c2 - b2 * c1) < 1e-12
        )
        if same_line:
            result_text = "Infinite solutions (both equations represent the same line)."
        else:
            result_text = "No solution (lines are parallel)."

    if status1 == "invalid" or status2 == "invalid":
        result_text = "Invalid equation: both coefficients of x and y cannot be zero for an equation."

    ax.axhline(0, color="black", linewidth=0.8)
    ax.axvline(0, color="black", linewidth=0.8)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.grid(True, linestyle="--", alpha=0.6)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("Interactive 2x2 Linear System")

    # Place legend outside plot area on the right.
    ax.legend(loc="upper left", bbox_to_anchor=(1.02, 1.0), borderaxespad=0.0)
    fig.subplots_adjust(right=0.72)

    plt.show()

    print("Equation 1:", f"{a1}x + {b1}y = {c1}")
    print("Equation 2:", f"{a2}x + {b2}y = {c2}")
    print(result_text)


# Interactive coefficient inputs
a1_w = widgets.FloatText(value=2.0, description="a1")
b1_w = widgets.FloatText(value=-3.0, description="b1")
c1_w = widgets.FloatText(value=-15.0, description="c1")

a2_w = widgets.FloatText(value=5.0, description="a2")
b2_w = widgets.FloatText(value=1.0, description="b2")
c2_w = widgets.FloatText(value=22.0, description="c2")

ui = widgets.VBox([
    widgets.HTML("<b>Equation 1:</b> a1x + b1y = c1"),
    widgets.HBox([a1_w, b1_w, c1_w]),
    widgets.HTML("<b>Equation 2:</b> a2x + b2y = c2"),
    widgets.HBox([a2_w, b2_w, c2_w]),
])

out = widgets.interactive_output(
    solve_and_plot,
    {
        "a1": a1_w,
        "b1": b1_w,
        "c1": c1_w,
        "a2": a2_w,
        "b2": b2_w,
        "c2": c2_w,
    },
)

display(ui, out)
```
