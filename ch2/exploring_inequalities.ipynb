# Inequalities with Reciprocals

## Overview
When we take reciprocals of inequalities, the direction of the inequality reverses **if and only if** both sides have the same sign. This notebook explores how $\frac{1}{x}$ relates to $\frac{1}{c}$ in different scenarios based on the signs and magnitudes of $x$ and $c$.

### Key Principle
For non-zero $x$ and $c$:
- If $x$ and $c$ have the **same sign**, taking reciprocals **reverses** the inequality
- If $x$ crosses zero (changes sign), we must consider the intervals separately

## Case 1: $x > c$ where $c > 0$

When $c$ is positive and $x > c$, both $x$ and $c$ are positive.

**Rule:** Taking reciprocals of positive numbers reverses the inequality.

$$x > c \implies \frac{1}{x} < \frac{1}{c}$$

**Example:** $x > 2 \implies \frac{1}{x} < \frac{1}{2}$
- If $x = 3$: $\frac{1}{3} \approx 0.33 < 0.5 = \frac{1}{2}$ ✓
- If $x = 5$: $\frac{1}{5} = 0.2 < 0.5 = \frac{1}{2}$ ✓
- As $x$ increases, $\frac{1}{x}$ decreases

## Case 2: $x < c$ where $c > 0$

When $c$ is positive and $x < c$, we need to consider two sub-cases:

### Sub-case 2a: $0 < x < c$
Both $x$ and $c$ are positive. Taking reciprocals reverses the inequality:

$$0 < x < c \implies \frac{1}{x} > \frac{1}{c}$$

**Example:** $0 < x < 2 \implies \frac{1}{x} > \frac{1}{2}$
- If $x = 1$: $\frac{1}{1} = 1 > 0.5$ ✓
- If $x = 0.5$: $\frac{1}{0.5} = 2 > 0.5$ ✓

### Sub-case 2b: $x < 0 < c$
Here $x$ is negative and $c$ is positive, so $\frac{1}{x}$ is negative and $\frac{1}{c}$ is positive.

$$x < 0 < c \implies \frac{1}{x} < 0 < \frac{1}{c}$$

**Example:** $x = -0.5, c = 2 \implies \frac{1}{x} = -2 < 0.5 = \frac{1}{c}$ ✓

## Case 3: $x > c$ where $c < 0$

### Sub-case 3a: $c < x < 0$

### Sub-case 3a: $c < x < 0$

$$c < x < 0 \implies \frac{1}{x} < \frac{1}{c}$$

**Example:** $-2 < x < 0 \implies \frac{1}{x} < \frac{1}{-2} = -0.5$

Here $x$ is positive and $c$ is negative, so $\frac{1}{x}$ is positive and $\frac{1}{c}$ is negative.

$$0 > c \text{ and } x > 0 \implies \frac{1}{c} < 0 < \frac{1}{x}$$

**Example:** $c = -2, x = 0.5 \implies \frac{1}{c} = -0.5 < 0.5 = \frac{1}{x}$ ✓

## Case 4: $x < c$ where $c < 0$

When both $c$ and $x$ are negative and $x < c$, both numbers are negative.

**Rule:** Taking reciprocals of negative numbers reverses the inequality direction.

$$x < c < 0 \implies \frac{1}{x} > \frac{1}{c}$$

**Example:** $x < -2 \implies \frac{1}{x} > \frac{1}{-2} = -0.5$
- If $x = -3$: $\frac{1}{-3} \approx -0.33 > -0.5$ ✓
- If $x = -5$: $\frac{1}{-5} = -0.2 > -0.5$ ✓
- As $x$ decreases (becomes more negative), $\frac{1}{x}$ increases (becomes less negative)

```python
import numpy as np
import matplotlib.pyplot as plt

# Setup for visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Reciprocal Inequalities: How 1/x relates to 1/c', fontsize=16, fontweight='bold')

# Case 1: x > c where c > 0
ax = axes[0, 0]
c1 = 2
x1_range = np.linspace(2.1, 10, 1000)
ax.plot(x1_range, 1/x1_range, 'b-', linewidth=2.5, label=f'1/x')
ax.axhline(y=1/c1, color='r', linestyle='--', linewidth=2, label=f'1/c = 1/{c1} = {1/c1:.2f}')
ax.axvline(x=c1, color='g', linestyle='--', linewidth=2, alpha=0.7, label=f'x = c = {c1}')
ax.fill_between(x1_range, 0, 1/x1_range, alpha=0.2, color='blue', label='Region: x > c')
ax.set_xlim(0, 10)
ax.set_ylim(-0.1, 1)
ax.set_xlabel('x', fontsize=11)
ax.set_ylabel('y', fontsize=11)
ax.set_title(f'Case 1: x > c ({c1})\n1/x < 1/c', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper right', fontsize=10)
ax.text(5, 0.3, 'As x increases\n1/x decreases', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Case 2: 0 < x < c where c > 0
ax = axes[0, 1]
c2 = 2
x2_range = np.linspace(0.1, c2-0.01, 1000)
ax.plot(x2_range, 1/x2_range, 'b-', linewidth=2.5, label=f'1/x')
ax.axhline(y=1/c2, color='r', linestyle='--', linewidth=2, label=f'1/c = 1/{c2} = {1/c2:.2f}')
ax.axvline(x=c2, color='g', linestyle='--', linewidth=2, alpha=0.7, label=f'x = c = {c2}')
ax.fill_between(x2_range, 1/x2_range, 10, alpha=0.2, color='blue', label='Region: 0 < x < c')
ax.set_xlim(0, 3)
ax.set_ylim(0, 10)
ax.set_xlabel('x', fontsize=11)
ax.set_ylabel('y', fontsize=11)
ax.set_title(f'Case 2: 0 < x < c ({c2})\n1/x > 1/c', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper right', fontsize=10)
ax.text(0.5, 5, 'As x decreases\n1/x increases', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Case 3: c < x < 0 where c < 0
ax = axes[1, 0]
c3 = -2
x3_range = np.linspace(c3+0.01, -0.1, 1000)
ax.plot(x3_range, 1/x3_range, 'b-', linewidth=2.5, label=f'1/x')
ax.axhline(y=1/c3, color='r', linestyle='--', linewidth=2, label=f'1/c = 1/{c3} = {1/c3:.2f}')
ax.axvline(x=c3, color='g', linestyle='--', linewidth=2, alpha=0.7, label=f'x = c = {c3}')
ax.fill_between(x3_range, 1/x3_range, 0, alpha=0.2, color='blue', label=f'Region: {c3} < x < 0')
ax.set_xlim(-3, 0)
ax.set_ylim(-10, 0)
ax.set_xlabel('x', fontsize=11)
ax.set_ylabel('y', fontsize=11)
ax.set_title(f'Case 3: {c3} < x < 0\n1/x < 1/c', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', fontsize=10)
ax.text(-2.7, -5, 'As x decreases\n(more negative)\n1/x decreases\n(more negative)', 
        fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Case 4: x < c where c < 0
ax = axes[1, 1]
c4 = -2
x4_range = np.linspace(-10, c4-0.1, 1000)
ax.plot(x4_range, 1/x4_range, 'b-', linewidth=2.5, label=f'1/x')
ax.axhline(y=1/c4, color='r', linestyle='--', linewidth=2, label=f'1/c = 1/{c4} = {1/c4:.2f}')
ax.axvline(x=c4, color='g', linestyle='--', linewidth=2, alpha=0.7, label=f'x = c = {c4}')
ax.fill_between(x4_range, 0, 1/x4_range, alpha=0.2, color='blue', label=f'Region: x < {c4}')
ax.set_xlim(-10, 0)
ax.set_ylim(-1, 0.1)
ax.set_xlabel('x', fontsize=11)
ax.set_ylabel('y', fontsize=11)
ax.set_title(f'Case 4: x < {c4}\n1/x > 1/c', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='lower left', fontsize=10)
ax.text(-8, -0.7, 'As x decreases\n1/x increases\n(closer to 0)', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```

## General Rule

**When the sign doesn't change** (both $x$ and $c$ are positive, or both are negative):
- Taking reciprocals **reverses** the direction of the inequality

**When the sign changes** (one is positive, one is negative):
- The reciprocal of the positive number is positive
- The reciprocal of the negative number is negative
- They cannot be directly compared through a simple inequality reversal

```python
from ipywidgets import interact, FloatSlider, IntSlider
import numpy as np

def explore_reciprocal_inequality(c_val, x_val):
    """
    Interactively explore how reciprocals relate to the original inequality.
    """
    print(f"Constants and Values:")
    print(f"  c = {c_val:.3f}")
    print(f"  x = {x_val:.3f}")
    print()
    
    # Check signs
    c_sign = "positive" if c_val > 0 else ("negative" if c_val < 0 else "zero")
    x_sign = "positive" if x_val > 0 else ("negative" if x_val < 0 else "zero")
    
    print(f"Signs: c is {c_sign}, x is {x_sign}")
    print()
    
    # Original inequality
    if x_val > c_val:
        print(f"Original: x > c → {x_val:.3f} > {c_val:.3f} ✓")
    elif x_val < c_val:
        print(f"Original: x < c → {x_val:.3f} < {c_val:.3f} ✓")
    else:
        print(f"Original: x = c → {x_val:.3f} = {c_val:.3f}")
    print()
    
    # Reciprocals
    if x_val != 0 and c_val != 0:
        recip_x = 1 / x_val
        recip_c = 1 / c_val
        
        print(f"Reciprocals:")
        print(f"  1/x = 1/{x_val:.3f} = {recip_x:.4f}")
        print(f"  1/c = 1/{c_val:.3f} = {recip_c:.4f}")
        print()
        
        # Check if reciprocal inequality reversed or held
        if x_val > c_val:
            if recip_x < recip_c:
                print(f"✓ Inequality REVERSED: 1/x < 1/c → {recip_x:.4f} < {recip_c:.4f}")
            elif recip_x > recip_c:
                print(f"✗ Inequality NOT reversed: 1/x > 1/c → {recip_x:.4f} > {recip_c:.4f}")
            else:
                print(f"Equal: 1/x = 1/c")
        elif x_val < c_val:
            if recip_x > recip_c:
                print(f"✓ Inequality REVERSED: 1/x > 1/c → {recip_x:.4f} > {recip_c:.4f}")
            elif recip_x < recip_c:
                print(f"✗ Inequality NOT reversed: 1/x < 1/c → {recip_x:.4f} < {recip_c:.4f}")
            else:
                print(f"Equal: 1/x = 1/c")
        print()
        
        # Analysis
        if (c_val > 0 and x_val > 0) or (c_val < 0 and x_val < 0):
            print("✓ Same sign: Reciprocal inequality reverses (as expected)")
        elif (c_val > 0 and x_val < 0) or (c_val < 0 and x_val > 0):
            print("✓ Different signs: 1/c and 1/x have opposite signs")
    else:
        print("⚠ Cannot take reciprocal of 0!")

# Create interactive sliders
interact(
    explore_reciprocal_inequality,
    c_val=FloatSlider(value=-2.0, min=-5, max=5, step=0.5, description='c'),
    x_val=FloatSlider(value=3.0, min=-5, max=5, step=0.1, description='x')
);
```

## Practice Problems & Solutions

### Problem 1
If $x > 3$, what can you conclude about $\frac{1}{x}$?

**Solution:**  
Since $x > 3$ and both $x$ and $3$ are positive, taking reciprocals reverses the inequality:
$$\frac{1}{x} < \frac{1}{3} \approx 0.333$$

### Problem 2
If $-4 < x < -1$, find the range of $\frac{1}{x}$.

**Solution:**  
Both $x$ and the bounds are negative. Taking reciprocals reverses the inequality:
$$-4 < x < -1 \implies \frac{1}{x} > \frac{1}{-1} = -1 \text{ and } \frac{1}{x} < \frac{1}{-4} = -0.25$$

Wait, let me be careful here. If $x = -2$, then $\frac{1}{x} = -0.5$. 
- When $x = -4$: $\frac{1}{x} = -0.25$
- When $x = -1$: $\frac{1}{x} = -1$
- When $x = -2$: $\frac{1}{x} = -0.5$

Since we're moving from $x = -4$ toward $x = -1$, and reciprocal reverses the inequality:
$$-0.25 > \frac{1}{x} > -1$$

Rewritten in standard form: $$-1 < \frac{1}{x} < -0.25$$

### Problem 3
Solve the inequality $\frac{1}{x} < 2$.

**Solution:**  
We need to consider cases based on the sign of $x$.

**Case A: $x > 0$**  
Since $x > 0$, multiply both sides by $x$ (inequality doesn't reverse):
$$1 < 2x \implies x > \frac{1}{2}$$

**Case B: $x < 0$**  
Since $x < 0$, multiply both sides by $x$ and reverse the inequality:
$$1 > 2x \implies x < \frac{1}{2}$$

But we're already in the case $x < 0$, so this is automatically satisfied.

**Final Answer:** $x < 0$ or $x > \frac{1}{2}$

### Problem 4
For what values of $x$ is $\frac{1}{x} > \frac{1}{5}$?

**Solution:**  
We need both sides to have the same sign to use the reciprocal reversal rule.

**Case A: Both positive** ($x > 0$ and $5 > 0$)  
Taking reciprocals reverses: $x < 5$  
Combined with $x > 0$: $0 < x < 5$

**Case B: Both negative** ($x < 0$ and... wait, $5 > 0$, so they don't have the same sign)

**Case B: $x < 0$ and $\frac{1}{5} > 0$**  
Here $\frac{1}{x} < 0$ always, and $\frac{1}{5} > 0$ always.  
So $\frac{1}{x} < 0 < \frac{1}{5}$, which means $\frac{1}{x} < \frac{1}{5}$ (NOT $>$).

**Final Answer:** $0 < x < 5$
