```python

import matplotlib.pyplot as plt
import numpy as np


plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 18})

n = 1000
q = int(n/4)
X = np.zeros((n,n))
X[(q-1):(3*q),(q-1):(3*q)] = 1

plt.imshow(X)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Unrotated Matrix')
plt.show()
```

It creates a 1000×1000 matrix of zeros, then sets a centered square region to 1.

- `n = 1000`: matrix size  
- `q = 250`: quarter of n  
- `X = np.zeros((n,n))`: 1000×1000 all zeros  
- `X[249:750, 249:750] = 1`: fills a 501×501 centered square with ones

So you get a white square on a black background when plotted.

```python
from scipy.ndimage import rotate

X_rot = rotate(X, 1, reshape=False, order=0)
X_rot[X_rot != 0] = 1

plt.imshow(X_rot)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Rotated Matrix')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices=0)


plt.semilogy(S,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Unrotated Matrix: Spectrum')
plt.show()
```


 `X_rot[X_rot != 0] = 1`:

**Boolean indexing:**
- `X_rot != 0` creates a boolean mask (True where values aren't zero)
- `X_rot[X_rot != 0]` selects only those elements
- `= 1` sets them all to 1

It finds every non-zero value and replaces it with 1.


```python

U_rot, S_rot, VT_rot = np.linalg.svd(X_rot,full_matrices=0)

print( S_rot.shape )

plt.semilogy(S_rot,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Rotated Matrix: Spectrum')
plt.show()

```
